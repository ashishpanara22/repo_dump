```python
x = df.loc[:,x_name]
y = y
```

```python
from sklearn import linear_model,tree,neighbors,ensemble,svm,model_selection
```

```python
x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,test_size=0.25,random_state=0)
```

Logistic regression

```python
mlr = linear_model.LogisticRegression()
mlr.fit(x_train,y_train)
```

```python
y_mlr = mlr.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,y_mlr)
```

```python
metrics.confusion_matrix(y_test,y_mlr)
```

```python
mdt = tree.DecisionTreeClassifier(random_state=0)
mdt.fit(x_train,y_train)
y_mdt = mdt.predict(x_test)
metrics.accuracy_score(y_test,y_mdt),metrics.confusion_matrix(y_test,y_mdt)
```

```python
y_mdt1 = mdt.predict(x_train)
metrics.accuracy_score(y_train,y_mdt1),metrics.confusion_matrix(y_train,y_mdt1)
```
